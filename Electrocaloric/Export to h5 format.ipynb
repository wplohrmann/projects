```python
def add_area(name,obj):
    #sample_area = 0.072938 #in square cm, 111
    #sample_area = 0.0133726 #in square cm, 011
    #sample_area = 0.014410  #in square cm, 001
    #if '.tab' not in name:
    #    return None
    if 'PMNPT111' in name:
        obj.attrs['Area / cm^2'] = 0.072938
    if 'PMNPT011' in name:
        obj.attrs['Area / cm^2'] = 0.0133726
    if 'PMNPT001' in name:
        obj.attrs['Area / cm^2'] = 0.014410
    if 'PLZT' in name and 'C.txt' in name:
        obj.attrs['Area / cm^2'] = np.pi * (50e-4)**2
    if 'Leakage' in name:
        obj.attrs['Area / cm^2'] = np.pi * (50e-4)**2
    
```

```python
def extractT(name):
    #pattern = r'([0-9]{3}\.[0-9]{2})'
    pattern = r'([0-9]{3})C'
    m = re.search(pattern,name)
    T_string = m.groups()[0]
    #T = int(name[-8:-4])
    return int(T_string)
    return int(T_string[:3]) + int(T_string[-2:])/100
def add_temperature(name,obj):
    #if '.tab' not in name:
    #    return None
    if 'Leakage Pvs t.txt' not in name:
        return None
    obj.attrs['Temperature'] = extractT(name)
```
