```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
arr = np.linspace(0,10,100)
```

```python
y = np.sin(arr)
```

```python
plt.plot(arr,y)
plt.show()
```

```python
data = np.transpose(np.vstack((arr,y)))
```

```python
data
```

```python
df = pd.DataFrame(data=data,columns=['x','y'])
```

```python
df
```

```python
df.to_csv('test_csv.csv',index=False,index_label=False)
```

```python
df_import = pd.read_csv('test_csv.csv')
```

```python
x = df_import['x']
y = df_import['y']
```

```python
plt.plot(x,y)
plt.show()
```
